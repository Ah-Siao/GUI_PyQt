### Set up the package

```python
!git clone https://github.com/SysBioChalmers/DLKcat
!pip install numpy requests torch torchvision rdkit-pypi scikit-learn
import os
import zipfile
import pandas as pd
dir='/content/DLKcat/DeeplearningApproach'
os.chdir(dir)
file=zipfile.ZipFile('Data/input.zip')
file.extractall('Data/')
file.close()
os.chdir('Code/example')

def customize_data(substrate,SMILES,Protein_Sequence):
    if type(substrate)==list:
        inputdic={'Substrate Name':substrate,'Substrate SMILES':SMILES,'Protein Sequence':Protein_Sequence}
    else:
        inputdic={'Substrate Name':[substrate],'Substrate SMILES':[SMILES],'Protein Sequence':[Protein_Sequence]}
    df=pd.DataFrame(inputdic)
    df.to_csv('inputs.tsv',sep='\t',index=False)
    return 'input file is created!'

def site_saturation_mutagenesis(substrate,SMILES,Protein_Sequence,site):
    inputdic={'Substrate Name':[],'Substrate SMILES':[],'Protein Sequence':[]}
    AA=['A','R','N','D','C','Q','E','G','H','I','L','K','M','F','P','S','T','W','Y','V']
    pos=site-1
    for i in AA:
        sub_name=f'{substrate}_{Protein_Sequence[pos]}{site}{i}'
        inputdic['Substrate Name'].append(sub_name)
        inputdic['Substrate SMILES'].append(SMILES)
        seq=list(Protein_Sequence)
        seq[pos]=i
        seq=''.join(seq)
        inputdic['Protein Sequence'].append(seq)
    df=pd.DataFrame(inputdic)
    df.to_csv('inputs.tsv',sep='\t',index=False)
    return 'The input file is created'

def double_mutation(substrate,SMILES,Protein_Sequence,site1,site2):
    inputdic={'Substrate Name':[],'Substrate SMILES':[],'Protein Sequence':[]}
    AA=['A','R','N','D','C','Q','E','G','H','I','L','K','M','F','P','S','T','W','Y','V']
    pos1=site1-1
    pos2=site2-1
    for i in AA:
      for j in AA:
        sub_name=f'{substrate}_{Protein_Sequence[pos1]}{site1}{i}_{Protein_Sequence[pos2]}{site2}{j}'
        inputdic['Substrate Name'].append(sub_name)
        inputdic['Substrate SMILES'].append(SMILES)
        seq=list(Protein_Sequence)
        seq[pos1]=i
        seq[pos2]=j
        seq=''.join(seq)
        inputdic['Protein Sequence'].append(seq)
    df=pd.DataFrame(inputdic)
    df.to_csv('inputs.tsv',sep='\t',index=False)
    return 'input table finished'

def triple_mutation(substrate,SMILES,Protein_Sequence,site1,site2,site3):
    inputdic={'Substrate Name':[],'Substrate SMILES':[],'Protein Sequence':[]}
    AA=['A','R','N','D','C','Q','E','G','H','I','L','K','M','F','P','S','T','W','Y','V']
    pos1=site1-1
    pos2=site2-1
    pos3=site3-1
    for i in AA:
      for j in AA:
        for k in AA:
          sub_name=f'{substrate}_{Protein_Sequence[pos1]}{site1}{i}_{Protein_Sequence[pos2]}{site2}{j}_{Protein_Sequence[pos3]}{site3}{k}'
          inputdic['Substrate Name'].append(sub_name)
          inputdic['Substrate SMILES'].append(SMILES)
          seq=list(Protein_Sequence)
          seq[pos1]=i
          seq[pos2]=j
          seq[pos3]=k
          seq=''.join(seq)
          inputdic['Protein Sequence'].append(seq)
    df=pd.DataFrame(inputdic)
    df.to_csv('inputs.tsv',sep='\t',index=False)
    return 'input table finished'
```

# Let users specify the mutation
Specify the amino acid which you'd like to change
Example: V461A; T101C; K232A (Remember to separate the words with ';')

M1A; Y2H; T3A

```python
Substrate='2-Oxopentanoate'
SMILES='CCCC(=O)C(=O)[O-]'
Protein_Sequence="MYTVGDYLLDRLHELGIEEIFGVPGDYNLQFLDQIISHKDMKWVGNANELNASYMADGYA\
RTKKAAAFLTTFGVGELSAVNGLAGSYAENLPVVEIVGSPTSKVQNEGKFVHHTLADGDF\
KHFMKMHEPVTAARTLLTAENATVEIDRVLSALLKERKPVYINLPVDVAAAKAEKPSLPL\
KKENSTSNTSDQEILNKIQESLKNAKKPIVITGHEIISFGLEKTVTQFISKTKLPITTLN\
FGKSSVDEALPSFLGIYNGTLSEPNLKEFVESADFILMLGVKLTDSSTGAFTHHLNENKM\
ISLNIDEGKIFNERIQNFDFESLISSLLDLSEIEYKGKYIDKKQEDFVPSNALLSQDRLW\
QAVENLTQSNETIVAEQGTSFFGASSIFLKSKSHFIGQPLWGSIGYTFPAALGSQIADKE\
SRHLLFIGDGSLQLTVQELGLAIREKINPICFIINNDGYTVEREIHGPNQSYNDIPMWNY\
SKLPESFGATEDRVVSKIVRTENEFVSVMKEAQADPNRMYWIELILAKEGAPKVLKKMGK\
LFAEQNKS"
```

```python
INPUT=input('Please specify the changes: ').split(';')
```

```python
import re
inputdic={'Substrate Name':[],'Substrate SMILES':[],'Protein Sequence':[]}
sub_name=f'{Substrate}_WT'
inputdic['Substrate Name'].append(sub_name)
inputdic['Substrate SMILES'].append(SMILES)
inputdic['Protein Sequence'].append(Protein_Sequence)

for i in INPUT:
  site=int(re.findall('\d+',i.strip())[0])
  pos=site-1
  mut=re.findall('[^0-9]',i.strip())[-1]
  sub_name=f'{Substrate}_{Protein_Sequence[pos]}{site}{mut}'
  inputdic['Substrate Name'].append(sub_name)
  inputdic['Substrate SMILES'].append(SMILES)
  seq=list(Protein_Sequence)
  seq[pos]=mut
  seq=''.join(seq)
  inputdic['Protein Sequence'].append(seq)

sub_name='_'.join(INPUT)
inputdic['Substrate Name'].append(sub_name)
inputdic['Substrate SMILES'].append(SMILES)
seq=list(Protein_Sequence)
for i in INPUT:
  site=int(re.findall('\d+',i.strip())[0])
  pos=site-1
  mut=re.findall('[^0-9]',i)[-1]
  seq[pos]=mut

seq=''.join(seq)
inputdic['Protein Sequence'].append(seq)

df=pd.DataFrame(inputdic)
df.to_csv('inputs.tsv',sep='\t',index=False)
print('Successfully rewrite the input table!')
```
